```python
# Cooling Rate Model, explanation for AE response
# Wang et al., 2022, Organic Geochemistry R2

import numpy as np
import matplotlib as mpl
from matplotlib import pyplot as plt
import pandas as pd
import os

plt.rcParams["font.family"] = ["Myriad Pro"]
plt.rcParams["font.size"] = 12
```

```python
## 50 and 0.5 C/Myr

d = {} # assign sheet names to dict

with open('CoolingModel_extract_0.5and50.xlsx', "rb") as xl:
     for sheet in ['SimTime', 'SurrT', 'IsoT', 'Keq', 'Delta', 'colors']:
            d[f'{sheet}'] = pd.read_excel(xl, sheet_name=sheet, header=[0,1])
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, gridspec_kw={'width_ratios': [1, 1]})
fig.tight_layout()
plt.subplots_adjust(wspace = 0.07)

for i in d['SimTime'][['Cooling Rate = 50 °C/Myr']].columns:
    ax1.plot(d['IsoT'][(i)], d['SurrT'][(i)], color=d['colors'].iloc[0][i], linestyle=d['colors'].iloc[1][i])
    initT = (np.round(d['SurrT'][(i)].iloc[0]))
    ax1.text(d['IsoT'][(i)].iloc[0], d['SurrT'][(i)].iloc[0], f' {initT:.0f} °C', 
             verticalalignment='center',
             horizontalalignment='left',
             fontsize=10, 
             rotation=0,
             color=d['colors'].iloc[0][i])
    finalT = (np.round(d['IsoT'][(i)].iloc[-1]))
    ax1.text(d['IsoT'][(i)].iloc[-1] + (1 if initT == 150 else 5 if initT == 125 else 0), d['SurrT'][(i)].iloc[-1], f'{finalT:.0f} °C', 
             verticalalignment='top',
             horizontalalignment='left' if initT == 125 else 'right',
             fontsize=10, 
             rotation=90,
             color=d['colors'].iloc[0][i])
    ax1.plot([80, 200], [80, 200], color='gray', linewidth=0.5, linestyle='dashed')
ax1.invert_yaxis()
ax1.set_xlim([70,295])
ax1.set_xlabel('Isotopic Temperature (°C)')
ax1.set_ylabel('Surrounding Temperature (°C)')
ax1.set_title('Cooling Rate = 50 °C/Myr', fontsize=10)
ax1.text(85, 78, 'A', 
         verticalalignment='top',
         fontweight='bold',
         fontsize=14, 
         color='black')
ax1.xaxis.grid(color='#DDDDDD', linestyle=':', linewidth=0.5)
ax1.yaxis.grid(color='#DDDDDD', linestyle=':', linewidth=0.5)
ax1.tick_params('both', length=5, width=1, which='major')
ax1.tick_params('both', length=1, width=1, which='minor')

for i in d['SimTime'][['Cooling Rate = 0.5 °C/Myr']].columns:
    ax2.plot(d['IsoT'][(i)], d['SurrT'][(i)], color=d['colors'].iloc[0][i], linestyle=d['colors'].iloc[1][i])
    initT = (np.round(d['SurrT'][(i)].iloc[0]))
    ax2.text(d['IsoT'][(i)].iloc[0], d['SurrT'][(i)].iloc[0], f' {initT:.0f} °C', 
             verticalalignment='center',
             horizontalalignment='left',
             fontsize=10, 
             rotation=0,
             color=d['colors'].iloc[0][i])
    finalT = (np.round(d['IsoT'][(i)].iloc[-1]))
    ax2.text(d['IsoT'][(i)].iloc[-1] + (5 if (initT == 175 or initT == 150) else 0), d['SurrT'][(i)].iloc[-1], f'{finalT:.0f} °C', 
             verticalalignment='top',
             horizontalalignment='left' if (initT == 175 or initT == 150) else 'right',
             fontsize=10, 
             rotation=90,
             color=d['colors'].iloc[0][i])
ax2.plot([80, 200], [80, 200], color='gray', linewidth=0.5, linestyle='dashed')
ax2.invert_yaxis()
ax2.yaxis.tick_right()
ax2.yaxis.set_label_position("right")
ax2.set_xlim([70,295])
ax2.set_xlabel('Isotopic Temperature (°C)')
ax2.set_ylabel('Surrounding Temperature (°C)')
ax2.set_title('Cooling Rate = 0.5 °C/Myr', fontsize=10)
ax2.text(85, 78, 'B', 
         verticalalignment='top',
         fontweight='bold',
         fontsize=14, 
         color='black')
ax2.xaxis.grid(color='#DDDDDD', linestyle=':', linewidth=0.5)
ax2.yaxis.grid(color='#DDDDDD', linestyle=':', linewidth=0.5)
ax2.tick_params('both', length=5, width=1, which='major')
ax2.tick_params('both', length=1, width=1, which='minor')
```

```python
fig.savefig('CoolingModelPlot_0.5and50.pdf', bbox_inches="tight") 
```
